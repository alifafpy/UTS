```python
# Melihat histogram dari Penderita diabetes berdasarkan jenis kelaminnya 
def histogram(kolom_data):
    perempuan = data[data['Sex'] == 1.0][kolom_data].value_counts()
    laki = data[data['Sex'] == 0.0][kolom_data].value_counts()
    dhist = pd.DataFrame([perempuan, laki])
    dhist.index = ['Perempuan', 'Laki-Laki']
    dhist.plot(kind = 'bar', stacked = True, figsize = (10, 5))
```

```python
histogram('Diabetes_binary')
```

# Normalisasi Data

Normalisasi Data digunakan untuk menormalisasikan data agar tidak ada data anomali sehingga data dapat berbentuk sebuah data yang lebih sederhana/standar, serta memiliki struktur yang lebih bagus.

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(data)
data_normalisasi = pd.DataFrame(d, columns = data.columns)
data_normalisasi.head(20)
```

```python
# Melakukan korelasi pada data yang sudah dinormalisasi
data_normalisasi.corr()
```

```python
# Melihat statistik deskriptif dari data yang sudah dinormalisasi
data_normalisasi.describe()
```

```python
fig, ax = plt.subplots(figsize = (15, 15))
sns.heatmap(data.corr(), ax = ax, annot = True)
```

```python
# Membagi data menjadi X dan y
X = data_normalisasi.drop(['Diabetes_binary'], axis = 1)
y = data_normalisasi['Diabetes_binary']
```

# Split dan Test Data

Di sini akan dilakukan pembagian pada data menjadi data Training dan data Testing.

```python
from sklearn.model_selection import train_test_split
```

```python
# Diambil data secara random untuk di test sebesar 30% dari dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

# Logistic Regression

Logistic Regression adalah metode yang digunakan untuk memprediksi nilai binary.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
```

```python
model_lr.fit(X_train, y_train)
```

```python
model_lr.score(X_test, y_test)
```

# KFold Logistic Regression

KFold adalah teknik yang digunakan untuk mengecek kevalidasian/keakuratan dari suatu model. Disini akan dicek keakuratan dari model berdasarkan metode Logistic Regression.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 0)
```

```python
score_lr = cross_val_score(model_lr, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_lr)
```

```python
score_lr.mean() * 100
```

Berdasarkan hasil perhitungan menggunakan metode Regresi Logistic, diperoleh tingkat akurasi sebesar 86%.

# K-Nearest Neighbors (KNN)

K-Nearest Neighbors (KNN) adalah sebuah metode yang digunakan untuk mengklasifikasikan suatu objek berdasarkan atribut dan sampel dari data training.

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors = 3)
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

# KFold KNN

KFold adalah teknik yang digunakan untuk mengecek kevalidasian/keakuratan dari suatu model. Disini akan dicek keakuratan dari model berdasarkan metode K_Nearest Neighbors (KNN).

```python
score_knn = cross_val_score(model_knn, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_knn)
```

```python
score_knn.mean() * 100
```

Berdasarkan hasil perhitungan menggunakan metode KNN, diperoleh tingkat akurasi sebesar 83%.

Dari 2 metode yang digunakan, yaitu metode Regresi Logistik dan KNN, diperoleh tingkat akurasi dari masing-masing model metode adalah sebesar 86% dan 83%. Berdasarkan hal tersebut, dapat dilihat bahwa tingkat akurasi dari metodel Regresi Logistik lebih tinggi, sehingga metode untuk model Regresi Logistik lebih baik dibanding KNN.
